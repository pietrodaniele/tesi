```python
from matplotlib import pyplot as plt
import uproot
import numpy as np
import itertools
import os
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
from dataclasses import dataclass, astuple, asdict

@dataclass(frozen=True) # classe
class RecoTrue:
    reco: str
    true: str
        
PARTICLES = "electron", "photon"
prefix = {'electron': 'el_', 'photon': 'ph_'}
```

```python
# Tutte le caratteristiche dei fotoni ed elettroni
BRANCHES_PH = ["ph_cl_E",
"ph_cl_eta",
#"ph_cl_etaCalo",
"ph_cl_phi",
#"ph_cl_phiCalo",
#"ph_convFlag",
"ph_Rconv",
"ph_pt1conv", "ph_pt2conv", "ph_ptconv",
"ph_zconv",
"ph_convtrk1nPixHits", "ph_convtrk1nSCTHits",
"ph_convtrk2nPixHits", "ph_convtrk2nSCTHits",
"ph_convtrk1InnermostPixelHits", "ph_convtrk1InnermostPixelExpHits", "ph_convtrk1NextToInnermostPixelHits", "ph_convtrk1NextToInnermostPixelExpHits",
"ph_convtrk2InnermostPixelHits", "ph_convtrk2InnermostPixelExpHits", "ph_convtrk2NextToInnermostPixelHits", "ph_convtrk2NextToInnermostPixelExpHits",                              
"index",
"ambiguos_index",
"actualIntPerXing",
"RunNumber",
"EventNumber",
"ph_truth_E",
"ph_truth_eta",
"ph_truth_pt",
"ph_truth_pdgId",
"ph_truth_parent_pdgId",
"ph_truth_matched",
"ph_is_ambiguous",
"ph_author",
"ph_amb_type",
"ph_truth_isConv",
"ph_truth_Rconv",
"ph_conv_type",
"el_truth_n",
"ph_truth_n",
"ph_minRfirstHit"               
#"npv",
#"ph_pass_tight"
              ]

BRANCHES_EL = ["el_cl_E",
"el_cl_eta",
"el_cl_etaCalo",
"el_cl_phi",
"el_cl_phiCalo",
"el_refittedTrack_qoverp",
"el_tracketa",
"el_trackpt",
"el_trackz0",
"el_author",
"index",
"ambiguos_index",
"actualIntPerXing",
"RunNumber",
"EventNumber",
"el_truth_E",
"el_truth_eta",
"el_truth_pt",
"el_truth_pdgId",
"el_truth_parent_pdgId",
"el_truth_matched",
"el_is_ambiguous",
"el_amb_type",
"el_truth_n",
"ph_truth_n",
"npv",
"el_pass_tight",
"el_trkPixelHits",
"el_trkSiHits",
"el_track_ep",
"el_trk_in_vertex",
"el_trkInnermostPixelHits", "el_trkInnermostPixelExpHits", "el_trkNextToInnermostPixelHits", "el_trkNextToInnermostPixelExpHits"
              ]
```

```python
base_dir = "data"
# definita una funzione che ricostruisce il filename dei dati che poi sarà passato a uproot.open
def get_filename(basedir, reco_true):
    fn = "output_ambiguity.true_%s.reco_%s.root" % (reco_true.true, reco_true.reco)
    return os.path.join(basedir, fn) # This method concatenates various path components with exactly one directory separator (‘/’)

def get_df(basedir, reco_true):
    f = uproot.open(get_filename(base_dir, reco_true))
    tree = f.get("egamma")
    branches = {"electron": BRANCHES_EL, "photon": BRANCHES_PH}[reco_true.reco]
    df = tree.pandas.df(branches)
    
    if reco_true.reco == 'photon':
        df['ph_isConv'] = (df['ph_Rconv'] > 0)# & (df['ph_Rconv'] < 800)
        
    return df
                    
all_reco_true = [RecoTrue(particle1, particle2) for particle1 in PARTICLES for particle2 in PARTICLES]

all_df = {reco_true: get_df(base_dir, reco_true) for reco_true in all_reco_true}
```

## Number of events

```python
df_nevents = []

for reco_true in all_reco_true:
    df = all_df[reco_true]
    k = asdict(reco_true)
    nevents = len(df['EventNumber'].unique())
    k.update({"size": nevents})
    df_nevents.append(k)
    
df_nevents = pd.DataFrame(df_nevents)
df_nevents = df_nevents.set_index(['reco', 'true'])['size'].unstack().T
df_nevents
```

## True energy distribution

```python
fig, ax = plt.subplots()
bins = np.linspace(0, 300E3, 50)
all_df[RecoTrue('electron', 'electron')]['el_truth_pt'].hist(bins=bins, histtype='step', label='true el', ax=ax)
all_df[RecoTrue('photon', 'photon')]['ph_truth_pt'].hist(bins=bins, histtype='step', label='true el', ax=ax)
ax.legend()
ax.set_xlabel('true pt')
```

## Join the two datasets
Let join the two dataset, only for the ambiguous particle, so that in the same dataframe we have the reco-particle of the two kind

### True photon
From the true photon AOD

```python
p = all_df[RecoTrue('photon', 'photon')].query('ph_is_ambiguous==True') # query verifica che i dati soddisfino una certa caratteristica
e = all_df[RecoTrue('electron', 'photon')].query('el_is_ambiguous==True')
df_true_photon = pd.merge(p, e, suffixes=('_ph', '_el'),
                          left_on=['EventNumber', 'ambiguos_index'],
                          right_on=['EventNumber', 'index'],
                          how='outer'
                         ) # merge unisce 
df_true_photon = df_true_photon.set_index(['EventNumber', 'index_ph'])
```

```python
df_true_photon[['index_el','ph_cl_E', 'el_cl_E', 'el_track_ep', 'ph_truth_pdgId', 'el_truth_pdgId', 'ph_truth_parent_pdgId', 'el_truth_parent_pdgId']].head(15)

# we are listing together the properties of the particle when recontructed as electron and
# when reconstructed as photon. Some properties as common (but maybe with different values since different
# calibration are applied) and some properties are not in common (e.g. track only for reco-electrons,
# conversion properties only for reco-photons)
```

### True electron

```python
p = all_df[RecoTrue('photon', 'electron')].query('ph_is_ambiguous==True')
e = all_df[RecoTrue('electron', 'electron')].query('el_is_ambiguous==True')
df_true_electron = pd.merge(e, p, suffixes=('_ph', '_el'),
                          right_on=['EventNumber', 'ambiguos_index'],
                          left_on=['EventNumber', 'index'])
df_true_electron = df_true_electron.set_index(['EventNumber', 'index_el'])

df_true_electron[['el_cl_E', 'ph_cl_E', 'el_track_ep', 'ph_truth_pdgId', 'el_truth_pdgId', 'ph_truth_parent_pdgId', 'el_truth_parent_pdgId']].head(15)

# probably we need to consider only the particle that are truth matched and with parent_pdgId = -999
```

## Everything in a single dataset

```python
df_amb = pd.concat({'el': df_true_electron.reset_index(), 'ph': df_true_photon.reset_index()}, names=['sample'])
df_amb = df_amb.reset_index().drop('level_1', axis=1)
df_amb.head(15)
```
